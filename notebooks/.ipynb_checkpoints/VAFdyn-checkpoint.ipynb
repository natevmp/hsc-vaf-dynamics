# VAF Clone Dynamics

```julia
include("../src/vafdyn.jl")
using .VAFDyn
using Plots
gr()
```

```julia
params = Dict(
    "ρ"=>1.,
    "μ"=>1.2,
    "N"=>100
)
```

## Evolution of a single clone

```julia
# discrete evolve
dfs = VAFDyn.DFreqspace(params["N"])
VAFDyn.addClones(dfs, Integer(round(params["N"]*0.3)), 1.)
VAFDyn.evolveVAF(dfs, params, 10, 0.001, addClones=false)

# diffusion evolve 
cfs = VAFDyn.CFreqspace(751)
VAFDyn.addClones(cfs, 0.3, 1.)
VAFDyn.evolveVAF(cfs, params, 10, 0.0001, addClones=false)

# kimura solution
probCloneKimura_f = VAFDyn.evolveCloneKimura(cfs.freqs_f, 0.3, 10, params["N"], 60);
```

```julia
# plotting
h = plot(dfs.freqs_f[2:end], dfs.n_f[2:end], label="MC", linewidth=3)
plot!(cfs.freqs_f[2:end-1], cfs.n_f[2:end-1]/params["N"], linestyle=:dash, label="PDE", linewidth=3)
plot!(cfs.freqs_f, probCloneKimura_f/params["N"], linestyle=:dashdot, label= "Kimura solution", linewidth=3)
xlims!((0, 1))
xlabel!("variant frequency")
ylabel!("probability")
title!("Discrete Freqspace")
display(h)
```
