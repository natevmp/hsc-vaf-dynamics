```julia
using Distributions
using LaTeXStrings
```

```julia
using Plots
gr()
```

```julia
ρ = 50

pois1 = Poisson(ρ)
pois2 = Poisson(2ρ)
```

```julia
x1 = 2*rand(pois1, 900000);
x2 = rand(pois2, 900000);
```

```julia
plot1 = histogram(x1, bins=100, normalize=true, alpha=0.6, label=L"2x; \; x \sim \mathrm{Pois}\, (\rho)")
histogram!(x2, bins=100, normalize=true, alpha=0.6, label=L"x; \; x \sim \mathrm{Pois}\, (2\rho)")

display(plot1)
```

```julia
plot2 = plot(pdf.(pois1, 0:100), label="2 Pois(ρ)")
# plot!(pdf.(pois2, range(50, 150; length=100)), label="Pois(ρ)")

display(plot2)
```
